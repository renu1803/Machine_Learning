# Linear Regression

## 1. Homeprice

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('/content/homeprices-onehot.csv')
data
```

```python
dummies=pd.get_dummies(data.town,dtype=int)
dummies
```

```python
merged = pd.concat([data, dummies], axis='columns')
merged
```

```python
final= merged.drop(['town','west windsor'],axis='columns')
final
```

```python
x=final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
model = LinearRegression()
model.fit(x, y)
```

**Build a predictor function to predict the price of a home.**
1. With 3400sq ft area in west windsor.
2. 2800sq ft area in robinsville


```python
import pandas as pd
model.predict(pd.DataFrame([[3400, 0, 0]], columns=['area', 'monroe township', 'robinsville']))
```

```python
predicted_price_robinsville = model.predict(pd.DataFrame([[2800, 0, 1]], columns=['area', 'monroe township', 'robinsville']))
print(f"Predicted price for a 2800 sq ft home in Robinsville: {predicted_price_robinsville[0]:,.2f}")
```

```python
predicted_price_Westwindsore = model.predict(pd.DataFrame([[3400, 0, 0]], columns=['area', 'monroe township', 'robinsville']))
print(f"Predicted price for a 3400 sq ft home in West windsore: {predicted_price_Westwindsore[0]:,.2f}")
```

```python
model.score(x, y)
```
