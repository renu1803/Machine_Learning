```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("student_clustering.csv")
print("The shape of the data is ",df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],df["iq"])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters=i)
  km.fit(df)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
X[y_means == 3,1 ]
```

```python
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],s=100,c="red",label="Cluster 1")
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],s=100,c="blue",label="Cluster 2")
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],s=100,c="green",label="Cluster 3")
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],s=100,c="yellow",label="Cluster 4")
plt.legend()
```

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]
X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
X
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0],y=X[:,1],z=X[:,2])
fig.show()
```

```python
wcss = []
for i in range(1,21):
  km = KMeans(n_clusters=i)
  km.fit_predict(X)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Assuming 2 clusters based on the make_blobs centers
km_2d = KMeans(n_clusters=2, random_state=1, n_init=10)
y_pred = km_2d.fit_predict(X)

df = pd.DataFrame()
df["x"] = X[:,0]
df["y"] = X[:,1]
df["cluster"] = y_pred
```

```python
import plotly.express as px

# Plotting the 2D clusters
fig = px.scatter(df, x="x", y="y", color="cluster")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Recreate df with 3D data and cluster predictions from 3D KMeans
df = pd.DataFrame()
df["x"] = X[:,0]
df["y"] = X[:,1]
df["z"] = X[:,2]
df["cluster"] = y_pred

# Plot the 3D clusters
fig = px.scatter_3d(df, x='x', y='y', z='z', color='cluster')
fig.show()
```
